<p>
En este ejercicio queremos diseñar un "solucionador" de sudokus utilizando estrategias de backtracking. Para ello se os da la clase Sudoku que hace las veces de modelo y contiene funciones útiles que os permitirán centraros en el desarrollo del algoritmo.
</p><p>
El problema del sudoku consiste, basicamente en rellenar un tablero con números del 1 al 9 (ambos incluidos). Para insertar los numeros de forma válida en el tablero se tienen que cumplir tres condiciones:
</p>
<ul>
    <li> El número no se encuentra repetido en la misma fila</li>
<li> El número no se encuentra repetido en la misma columna</li>
<li> El número no se encuentra repetido en el mismo bloque*</li>
</ul>
<p>
*: Un bloque es cada una de las sub-divisiones 3x3 del tablero.</p>
</div>

<center><p><h3> Código </h3></p> </center> 

<p>
<h3>INPUT</h3>
<ul>
<li>s: instancia de la clase Sudoku</li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
    <li>s: misma instancia de la clase Sudoku resuelto</li>
    <li>expanded: número de estados explorados hasta encontrar la solución</li>
<ul>

````python
from sudoku import Sudoku

# Para cargar el sudoku introducimos la ruta al archivo. En la carpeta './sudokus' tenéis un buén número
# de ejemplos con los que podéis jugar

print("1- Crear un sudoku")
s2 = Sudoku('sudoku/s2')
print(s2)

# Dada una posición definida como una tupla de dos elementos (x,y) en el rango [0,8], podemos obtener los
# elementos validos para una posición vacía

print("2- Obtén los números válidos en (0,1)")
print(s2.get_valid((0,1)))

# En caso que se trate de una posición que ya contiene un número devolverá una lista vacía
print(s2.get_valid((0,0)))

# También podemos obtener la lista de todos los elementos vacios
print("3 - Obtén las posiciones vacías")
print(s2.get_empty())

# Finalmente, nos permite insertar un numero en una posición dada.
print("4 - Añade un 7 a la posición (0,1)")
s3 = s2.set_number((0,1),7)
print(s2)
print(s3)

# En caso de que no se pueda insertar el numero en esa posición nos devolverá un mensaje de error:
_set = s2.set_number((0,1),2)
print(_set)
if _set:
    print('shit')
if not _set:
    print('no shit')

print(s3.get_empty())
````

````python
def sudoku_solver(s):
    """
    Función que resuelve el sudoku mediante backtracking. El retorno es el mismo objeto Sudoku solucionado.
    """
    _s_empty = s.get_empty()
      
    # If there is no unassigned location, we are done     
    if(not _s_empty): 
        return (True, s)
      
    # get the first empty item  
    row = _s_empty[0][0] 
    col = _s_empty[0][1] 
      
    # consider digits 1 to 9 
    for num in range(1,10):
        
        #check if the possition admits more valid numbers
        if (len(s.get_valid((row,col)))>0):
            
            #now check if the number is one of the valids
            if (num in s.get_valid((row,col))):

                # make the assignment if its valid
                _set = s.set_number((row,col),num)
                if _set:
                    
                    _sol_s = sudoku_solver(_set)
                    if(_sol_s[0]):
                        # return that it have succed and the solution
                        return (True, _sol_s[1])
              
    # return that the given sudoku it's not solvable        
    return (False, 'Can\'t sove it')
````

````python
sudoku = Sudoku('sudoku/s3')
print(sudoku)
sol = sudoku_solver(sudoku)
print(sol[1])
````

Se os pide que programéis una función que sea capaz de resolver un 8-Puzzle dada cualquier posible configuración inicial (La función NPuzzle.create_board solo devuelve configuraciones que tengan solución)

La implementación del algoritmo ha de ser utilizando ramificación y poda. Para ramificar el árbol de estados utilizaremos como heurística la suma de la distáncia de Manhattan de todas las piezas a su posición objetivo. Esta función se os da implementada de la siguiente forma: **NPuzzle.manhattan_distance(board)**

El objetivo del código es encontrar la solución óptima, que en este caso es la que tiene un menor número de pasos. Por lo tanto, para la implementación de la poda, utilizaremos como valor de cota la profundidad.

<div class="alert alert-danger" style="width:80%; margin:0 auto; padding">
<center><p><h3> Código </h3></p> </center> 

<p>
<h3>INPUT</h3>
<ul>
<li>board: instancia de la clase NPuzzle</li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
    <li>distance: número de movimientos realizados para resolver el puzzle</li>
    <li>final_board: misma instancia de la clase NPuzzle resuelto</li>
    <li>expanded: número de estados explorados hasta encontrar la solución</li>
<ul>

</p>

````python
%reset -f
from npuzzle import NPuzzle
from heapq import heappop, heappush

def solve_puzzle(board):
    
    expanded = 0
    
    #Create a priority queue to store the "opened" nodes, node = (heur, dist, ocurrance, board)
    
    #it's important to have the occurrance in the node so in case the cost and the dist from root
    #are equal between nodes, heapq can't compare Npuzzels, so we have to provide an element to
    #dicide the order in these cases. And as we are looking for the solotion with smallest number of moves
    #it's better this order, (cost+dist, dist, ocurrance, board), because dist = moves from origin is more
    #important than in which order we found the possible solution.
    
    Q = []
    
    #Take the inicial state as the root, calculate its cost
    #and put it in the queue using heapq
    heappush(Q, (board.manhattan_distance(), 0, expanded, board))
    expanded += 1
    
    while len(Q)>0:
        
        #Take the element with smallest cost in the queue
        min_node = heappop(Q)

        #if it's the answer, return it
        if min_node[3].state():
            return min_node[0], min_node[3], expanded

        for mov in min_node[3].allowed_moves():
            #push the next nodes fund by making the possible moves
            new_board = min_node[3].move(mov)
            
            heappush(Q, (new_board.manhattan_distance()+min_node[1]+1, min_node[1]+1, expanded, new_board))
            expanded += 1
    
    return 0, board, expanded
````

````python
from npuzzle import NPuzzle
from heapq import heappop, heappush

#As some times it takes to long in order to converge to the solution we use exp to no do too many operations
#and we get exp from the function above
def solve_puzzle2(board, exp):
    
    expanded = 0
    
    #Create a priority queue to store the "opened" nodes, node = (cost, dist, ocurrance, board)
    
    #it's important to have the occurrance in the node so in case the cost and the dist from root
    #are equal between nodes, heapq can't compare Npuzzels, so we have to provide an element to
    #dicide the order in these cases. And as we are looking for the solotion with smallest number of moves
    #it's better this order, (cost+dist, dist, ocurrance, board), because dist = moves from origin is more
    #important than in which order we found the possible solution.
    
    Q = []
    
    #Take the inicial state as the root, calculate its cost
    #and put it in the queue using heapq
    heappush(Q, (board.manhattan_distance(), 0, expanded, board))
    expanded += 1
    
    while len(Q)>0 and expanded<=exp*100:
        
        #Take the element with smallest cost in the queue
        min_node = heappop(Q)

        #if it's the answer, return it
        if min_node[0] == 0:
            return min_node[1], min_node[3], expanded

        for mov in min_node[3].allowed_moves():
            #push the next nodes fund by making the possible moves
            new_board = min_node[3].move(mov)
            
            heappush(Q, (new_board.manhattan_distance(), min_node[1]+1, expanded, new_board))
            expanded += 1
    
    return 0, board, expanded
````

I created two functions, the only difference between one and other is the heuristic used.

In the first one I use h(n)=f(n)+g(n), where f(n) are the number of steps done to get to the current node and g(n) it's the sum of manhattan distance of all the postions and the target postion.

In the second I use just g(n), which is the sum of manhattan distance of all the postions and the target postion.

I turns out that the first one some times needs a bit more steps compared with the second one, but not too many. But there are cases when the second one take TOO much to get to the solution, that can be caused cause of a step where the heuristics make the function go back and forth to the same two nodes, while the first avoids this by increasing the heuristic function every step done.
