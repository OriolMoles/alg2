<div class="alert alert-success" style="width:90%; margin:0 auto;">
  <h2><p>Sudoku</p></h2>
<p>
En este ejercicio queremos diseñar un "solucionador" de sudokus utilizando estrategias de backtracking. Para ello se os da la clase Sudoku que hace las veces de modelo y contiene funciones útiles que os permitirán centraros en el desarrollo del algoritmo.
</p><p>
El problema del sudoku consiste, basicamente en rellenar un tablero con números del 1 al 9 (ambos incluidos). Para insertar los numeros de forma válida en el tablero se tienen que cumplir tres condiciones:
</p>
<ul>
    <li> El número no se encuentra repetido en la misma fila</li>
<li> El número no se encuentra repetido en la misma columna</li>
<li> El número no se encuentra repetido en el mismo bloque*</li>
</ul>
<p>
*: Un bloque es cada una de las sub-divisiones 3x3 del tablero.</p>
</div>

<div class="alert alert-danger" style="width:80%; margin:0 auto; padding">
<center><p><h3> Código </h3></p> </center> 

<p>
<h3>INPUT</h3>
<ul>
<li>s: instancia de la clase Sudoku</li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
    <li>s: misma instancia de la clase Sudoku resuelto</li>
    <li>expanded: número de estados explorados hasta encontrar la solución</li>
<ul>

</p>

</div>

````python
from sudoku import Sudoku

# Para cargar el sudoku introducimos la ruta al archivo. En la carpeta './sudokus' tenéis un buén número
# de ejemplos con los que podéis jugar

print("1- Crear un sudoku")
s2 = Sudoku('sudoku/s2')
print(s2)

# Dada una posición definida como una tupla de dos elementos (x,y) en el rango [0,8], podemos obtener los
# elementos validos para una posición vacía

print("2- Obtén los números válidos en (0,1)")
print(s2.get_valid((0,1)))

# En caso que se trate de una posición que ya contiene un número devolverá una lista vacía
print(s2.get_valid((0,0)))

# También podemos obtener la lista de todos los elementos vacios
print("3 - Obtén las posiciones vacías")
print(s2.get_empty())

# Finalmente, nos permite insertar un numero en una posición dada.
print("4 - Añade un 7 a la posición (0,1)")
s3 = s2.set_number((0,1),7)
print(s2)
print(s3)

# En caso de que no se pueda insertar el numero en esa posición nos devolverá un mensaje de error:
s2.set_number((0,1),2)
````

````python
def sudoku_solver(s):
    """
    Función que resuelve el sudoku mediante backtracking. 
    El retorno es el mismo objeto Sudoku solucionado.
    
    Parameters
    ----------
        s, sudoku.Sudoku
            Sudoku object to solve.
        
    Returns
    -------
        boolean
            If the sudoku has been solved
        
        sudoku.Sudoku
            Solved sudoku object.
    """
    
    # Get the empty positions of 
    # the current sudoku
    
    empty_positions = s.get_empty()
      
    # Base case:
    # If there's no empty positions,
    # the sudoku is filled, i.e. solved
    
    if len(empty_positions) == 0: 
        return True, s
      
    # Recursive step:
    # We will take the first empty position
    # in the list, and try to fill it
    # with a number from 1 to 9
    
    pos = empty_positions[0]
    
    for candidate in range(1, 10):
        
        # Compute the valid numbers to fill the 
        # position
        
        valid_candidates = s.get_valid(pos)
        
        # If there's no valid number to fill it,
        # consider the next candidate
        
        if len(valid_candidates) == 0:
            continue
            
        # Check if the candidate is one of the valid
        # number to fill, if not consider the next candidate

        if candidate not in valid_candidates:
            continue

        # Try to set the position with the candidate
        
        s_updated = s.set_number(pos, candidate)
        
        if s_updated is None:
            continue

        # Make recursive call with the new sudoku
            
        is_solved, solved_sudoku = sudoku_solver(s_updated)

        if is_solved:
            return is_solved, solved_sudoku
              
    # Return False, None if we could not solve the sudoku
    
    return False, None
````

````python
from time import time

sudoku = Sudoku('sudoku/s3')

t = time()
is_solved, sudoku_solved = sudoku_solver(sudoku)
t = time() - t

print(f"Could solve:  {is_solved}")
print(f"Time elasped: {t:4.6f} s.")

if is_solved:
    print(sudoku_solved)
````

Se os pide que programéis una función que sea capaz de resolver un 8-Puzzle dada cualquier posible configuración inicial (La función NPuzzle.create_board solo devuelve configuraciones que tengan solución)

La implementación del algoritmo ha de ser utilizando ramificación y poda. Para ramificar el árbol de estados utilizaremos como heurística la suma de la distáncia de Manhattan de todas las piezas a su posición objetivo. Esta función se os da implementada de la siguiente forma: **NPuzzle.manhattan_distance(board)**

El objetivo del código es encontrar la solución óptima, que en este caso es la que tiene un menor número de pasos. Por lo tanto, para la implementación de la poda, utilizaremos como valor de cota la profundidad.

<div class="alert alert-danger" style="width:80%; margin:0 auto; padding">
<center><p><h3> Código </h3></p> </center> 

<p>
<h3>INPUT</h3>
<ul>
<li>board: instancia de la clase NPuzzle</li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
    <li>distance: número de movimientos realizados para resolver el puzzle</li>
    <li>final_board: misma instancia de la clase NPuzzle resuelto</li>
    <li>expanded: número de estados explorados hasta encontrar la solución</li>
    <li>moves: lista de movimientos realizados para resolverlo.</li>
<ul>

</p>

</div>

````python
%reset -f

from npuzzle import NPuzzle
from heapq import heappush as push
from heapq import heappop as pop

def solve_puzzle(board):
    """
    Solve a NPuzzle.
    
    Parameters
    ----------
        board, npuzzle.NPuzzle
        
    Returns
    -------
        distance, int
            Number of movements done
            
        final_board, npuzzle.NPuzzle
        
        expanded, int
            Number of explored states while solving
            
        moves, list
            List of done movements
        
    """
    
    queue, distance, expanded, moves = [], 0, 0, []
    
    # We use a heap queue structure so we can 
    # pop the best node candidate based on the cost to get there.
    # The cost function is defined by 
    #    C(b) = MD(b) + ED(b)
    # Where 
    #   b:  Node board
    #   MD: Manhattan distance of b to the solution
    #   ED: Distance from input board (root)
    
    push(queue, 
         (
            board.manhattan_distance(), 
            distance, 
            board, 
            expanded, 
            moves
        ))
    
    expanded += 1
    
    while len(queue) > 0:
        
        # Use the heap queue class to pop the 
        # element with smallest cost
        
        _, e_distance, e_board, e_expanded, e_moves = pop(queue)
        
        # Similar to the sudoku algorithm:
        # If the expanded board is solved, return 
        
        if e_board.state():
            return e_distance, e_board, e_expanded, e_moves

        # Now, as before, loop over all allowed moves for
        # the expanded board and add them to the heap queue.
        
        for a_movement in e_board.allowed_moves():
            
            # Make the movement to the expanded board and
            # add a new element corresponding to this new node
            
            new_board = e_board.move(a_movement)
            
            # Add the current movement
            
            new_moves = e_moves.copy()
            new_moves.append(a_movement)
            
            # Note the first element of the noded added (cost)
            # Iis the sum of the man. distance and the distance
            # from the root node
            
            push(queue, 
                 (
                    new_board.manhattan_distance() + e_distance + 1,
                    e_distance + 1, 
                    new_board,
                    expanded,
                    new_moves
                ))
            
            # We expanded one more node
            
            expanded += 1

    # If we end up here
    # then we could not solve, the board.
    
    return distance, board, expanded, moves
````

````python
from time import time

board = NPuzzle()
board.create_board(moves=10, n=7)

print("Board to solve")
print(board)

t = time()
distance, final_board, expanded, moves = solve_puzzle(board)
t = time() - t

print (f"Elapsed:  {t:4.6f} s")
print (f"Distance: {distance}")
print (f"Expanded: {expanded}")
print (f"Moves:    {moves}")
print (final_board)
````
